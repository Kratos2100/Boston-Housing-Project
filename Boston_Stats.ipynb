```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(rc = {'figure.figsize':(12,15)})
```

```python
boston_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ST0151EN-SkillsNetwork/labs/boston_housing.csv'
boston_df = pd.read_csv(boston_url)
```

```python
boston_df[0:20]
```

```python
medv = sns.boxplot(y = 'MEDV', data = boston_df)
```

```python
chas = pd.get_dummies(boston_df[['CHAS']])
```

```python
plot = sns.barplot(y = 'CHAS' , data = chas)
```

```python
air_quality = sns.scatterplot(x = 'INDUS', y = 'NOX' , hue = 'NOX', data = boston_df)
```

```python
pt_ratio = plt.hist(boston_df['PTRATIO'])
```

```python
boston_df.loc[(boston_df['AGE'] <= 35), 'AGE_GROUP'] = '35 and younger'
boston_df.loc[(boston_df['AGE'] > 35)&(boston_df['AGE'] < 70), 'AGE_GROUP'] = 'between 35 and 70'
boston_df.loc[(boston_df['AGE'] >=70), 'AGE_GROUP'] = '70 and above'

boxplot = sns.boxplot(x = 'AGE_GROUP', y = 'MEDV', data = boston_df)
boxplot.set(xlabel = 'AGE', ylabel = 'MEDV', title = 'MEDV vs AGE')
```
